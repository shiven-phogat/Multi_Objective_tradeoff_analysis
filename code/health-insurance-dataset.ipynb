```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import optuna
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load data
df = pd.read_csv("/kaggle/input/health-insurance-cross-sell-prediction/train.csv")

# Fill missing values
df['Gender'].fillna('Male', inplace=True)
df['Vehicle_Age'].fillna('> 2 Years', inplace=True)
df['Vehicle_Damage'].fillna('Yes', inplace=True)

# Encode categorical features
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df['Vehicle_Damage'] = df['Vehicle_Damage'].map({'Yes': 1, 'No': 0})
df = pd.get_dummies(df, columns=['Vehicle_Age'], drop_first=True)

# Feature selection
features = ['Gender', 'Age', 'Driving_License', 'Region_Code', 'Previously_Insured',
            'Vehicle_Damage', 'Annual_Premium', 'Policy_Sales_Channel', 'Vintage']+[col for col in df.columns if 'Vehicle_Age' in col]
X = df[features]
y = df['Response']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train_scaled, X_test_scaled, y_train, y_test, gender_train, gender_test = train_test_split(
    X_scaled, y, df['Gender'], test_size=0.2, random_state=42, stratify=y)

# Fairness metric: DPD
def compute_dpd(y_pred, gender):
    p_male = y_pred[gender == 1].mean()
    p_female = y_pred[gender == 0].mean()
    return abs(p_male - p_female)

# Objective function for Optuna
def objective(trial, X_train, y_train, X_val, y_val, gender_val, alpha):
    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'use_label_encoder': False,
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.3, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 20),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),
        'random_state': 42
    }
    model = xgb.XGBClassifier(**params)
    model.fit(X_train, y_train, verbose=False)
    preds = (model.predict_proba(X_val)[:, 1] >= 0.5).astype(int)
    acc = accuracy_score(y_val, preds)
    dpd = compute_dpd(preds, gender_val)
    return alpha * (1 - acc) + (1 - alpha) * dpd

# Adaptive Alpha Optimization
initial_alphas = np.linspace(0, 1, 15).tolist()
results = []

X_t, X_v, y_t, y_v, gender_t, gender_v = train_test_split(
    X_train_scaled, y_train, gender_train, test_size=0.25, random_state=123, stratify=y_train)

def optimize_for_alpha(alpha):
    print(f"\nüîç Optimizing for alpha = {alpha:.4f}")
    study = optuna.create_study(direction="minimize")
    study.optimize(lambda trial: objective(trial, X_t, y_t, X_v, y_v, gender_v, alpha), n_trials=100)

    best_model = xgb.XGBClassifier(**study.best_params, use_label_encoder=False, objective='binary:logistic', random_state=42)
    best_model.fit(X_train_scaled, y_train)
    preds = (best_model.predict_proba(X_test_scaled)[:, 1] >= 0.5).astype(int)
    acc = accuracy_score(y_test, preds)
    dpd = compute_dpd(preds, gender_test)
    print(f"‚úÖ Alpha {alpha:.4f}: Accuracy = {acc:.4f}, DPD = {dpd:.4f}")
    results.append((dpd, acc, alpha, study.best_params))

# Initial optimization
for alpha in initial_alphas:
    optimize_for_alpha(alpha)

# Adaptive refinement
max_iterations = 25
min_points = 20
epsilon = 0.01
iteration = 0

while iteration < max_iterations and len(results) < min_points:
    results.sort(key=lambda x: x[2])  # sort by alpha
    alphas_done = [r[2] for r in results]
    new_alphas = []

    for i in range(len(results) - 1):
        a1, a2 = results[i][2], results[i+1][2]
        d1, d2 = results[i][0], results[i+1][0]
        if abs(d2 - d1) > epsilon:
            mid_alpha = round((a1 + a2) / 2, 4)
            if mid_alpha not in alphas_done and mid_alpha not in new_alphas:
                new_alphas.append(mid_alpha)

    if not new_alphas:
        break

    for alpha in new_alphas:
        optimize_for_alpha(alpha)

    iteration += 1

# Sort results by DPD
results.sort(key=lambda x: x[0])

# Strict Pareto Frontier
def pareto_frontier(points):
    pareto = []
    for i, (dpd_i, acc_i) in enumerate(points):
        if not any(
            (acc_j >= acc_i and dpd_j <= dpd_i) and (acc_j > acc_i or dpd_j < dpd_i)
            for j, (dpd_j, acc_j) in enumerate(points) if i != j
        ):
            pareto.append((dpd_i, acc_i))
    return np.array(pareto)

pareto_points = [(dpd, acc) for (dpd, acc, _, _) in results]
frontier = pareto_frontier(pareto_points)

plt.figure(figsize=(6, 5))
plt.plot(frontier[:, 0], frontier[:, 1], 'r-o', label='Pareto Frontier')
plt.xlabel("DPD")
plt.ylabel("Accuracy")
plt.title("Pareto Frontier (Fairness vs. Accuracy)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
